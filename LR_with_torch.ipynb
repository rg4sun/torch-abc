```python
import torch
```

LR 模型： $ \hat{y} = w\cdot x + b $

```python
# 使用torch创建数据集
x_data = torch.Tensor([[1.], [2.], [3.]])
y_data = torch.Tensor([[2.], [4.], [6.]])
```

```python
# 构建模型
class LinearModel(torch.nn.Module): # 创建自己的模型，继承自nn.Module
    def __init__(self):# 类的构造函数
        super(LinearModel, self).__init__() # 调用父类的构造函数
        self.linear = torch.nn.Linear(1, 1) # 构造Linear对象，包含 weights 和 bias两个参数
        # torch.nn.Linear(in_features=1, out_features=1, bias=True) 输入样本规模，输出样本规模，偏置默认有偏置
    
    def forward(self, x):
        y_pred = self.linear(x)
        return y_pred
```

```python
model = LinearModel()
model
```

```python
# 构建loss函数
criterion = torch.nn.MSELoss(size_average=False) # 不对损失求均值，就是1/N
optimizer = torch.optim.SGD(model.parameters(), lr=0.01) # model.parameters()获取模型需要优化的参数（此处即w），lr为学习率固定为0.01
criterion
```

```python
# 构建loss函数
criterion = torch.nn.MSELoss(reduction='sum') # 不对损失求均值，就是1/N
criterion
```

```python
optimizer = torch.optim.SGD(model.parameters(), lr=0.01) # model.parameters()获取模型需要优化的参数（此处即w），lr为学习率固定为0.01
optimizer
```

```python
# 训练过程
for epoch in range(100):
    y_pred = model(x_data) 
    loss = criterion(y_pred, y_data)
    # print('Epoch:', epoch, 'loss=', loss, 'w=', model.linear.weight.item(), 'b=', model.linear.bias.item())
    # 上面的老版本了
    print('Epoch:', epoch, 'loss=', loss.item(), 'w=', model.linear.weight.item(), 'b=', model.linear.bias.item())

    optimizer.zero_grad() # 防止保存的梯度自动累加，这里对梯度进行每轮归零
    loss.backward() # 反向传播计算梯度
    optimizer.step() # 更新梯度
```

```python
x_test = torch.Tensor([[4.]])
y_test_pred = model(x_test)
y_test_pred
```
