```python
import torch as tc
```

```python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]

w = tc.Tensor([1.0])
w
```

```python
w.requires_grad = True # w变量设置为需要计算梯度
w
```

```python
def forward(x):
    return x*w # w为Tensor类型，*被重载，x会被强制转化成Tensor

def loss(x, y):
    y_pred = forward(x)
    return (y_pred - y)**2

# def gradient(x, y):
#     return 2*x*(x*w - y)
```

```python
print('Predict (before training)', 4, forward(4).item())
alpha = 0.1
for epoch in range(100):
    for x,y in zip(x_data, y_data):
        l = loss(x, y) # forward，只需要计算出loss;此步会创建计算图
        l.backward() # 直接调用torch自带的backward函数；自动计算并保存需要梯度的地方，此处只有w；完成backward会释放计算图
        print('\tgrad:', x, y, w.grad.item())
        w.data = w.data - alpha * w.grad.data
        # w 梯度置零
        w.grad.data.zero_()
        print('Epoch: ', epoch, 'loss=', l.item())
print('Predict (after training)', 4, forward(4).item())
```
